```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 21, dtype=np.float64)
```

```python
noise = np.random.normal(size=20)
```

```python
y = x + noise
```

```python
with plt.xkcd():
    plt.xticks([])
    plt.yticks([])
    plt.plot(x, y, label='Personal Growth')
    plt.plot(x, np.polyval(np.polyfit(x, y, 2), x), 'g--')
    plt.xlabel('Time')
    plt.ylabel('Skills')
    plt.legend()
```

```python
plt.show()
```

```python
plt.savefig('growth.pdf')
```
